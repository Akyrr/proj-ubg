```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset_preprocessing_raw.csv', na_values=['#','-'])
```

```python
df.head(50)
```

```python
df.info()
```

```python
df = df.drop(['customer_id'], axis=1)
```

```python
df['name'] = df['name'].replace('Joko#', 'Joko')
df['name'] = df['name'].str.title()
```

```python
df['age'] = df['age'].replace(200, np.nan)
med = df['age'].median()
df['age'] = df['age'].fillna(med)
```

```python
df['gender'] = df['gender'].map({'F':1, 'female':1, 'f':1, 'FEMALE':1, 'F':1, 'Male':0, 'M':0, '?':2})
```

```python
df['join_daten'] = pd.to_datetime(df['join_date'], errors='coerce')
df['join_date'] = df['join_daten'].dt.strftime('%y-%m-%d')
```

```python
# ga efektif kalo datasetnya kecil pake z score aja

medi = df['age'].median()
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)

IQR = Q3 - Q1

atas = Q1 + 1.5 * IQR
bawah = Q3 - 1.5 * IQR

df.loc[(df['age'] < bawah) | (df['age'] > atas)] = med
```
