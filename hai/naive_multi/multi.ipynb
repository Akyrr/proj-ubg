```python


# Train a simple (Bernoulli-style) Naive Bayes from the existing df and expose predict_text, nama_model, warna_model

# expect df to have a 'label' column and the rest are token/features
feature_cols = [c for c in df.columns if c != 'label']
X = df[feature_cols].values
y = df['label'].values
classes = np.unique(y)

# Train: class priors and per-feature probabilities with Laplace (alpha=1)
class_prior = {}
feature_prob = {}
alpha = 1.0

for c in classes:
    Xc = X[y == c]
    Nc = Xc.shape[0]
    class_prior[c] = Nc / len(y)
    count_j = Xc.sum(axis=0)  # counts of feature presence in class
    # Bernoulli likelihood: P(feature=1 | class)
    p_j = (count_j + alpha) / (Nc + 2 * alpha)
    feature_prob[c] = p_j

# Prepare name & color maps (customize as needed)
colors = ["#E74C3C", "#3498DB", "#2ECC71", "#9B59B6", "#F1C40F", "#1ABC9C", "#E67E22"]
nama_model = {c: f"Model {c}" for c in classes}
warna_model = {c: colors[i % len(colors)] for i, c in enumerate(classes)}

# Predictor used by the GUI
def predict_text(text):
    """
    Tokenize text and predict label.
    Returns the predicted class value (same type as df['label'] entries).
    """
    tokens = set(re.findall(r"\w+", text.lower()))
    # binary feature vector: 1 if column token appears in input
    x = np.array([1 if col.lower() in tokens else 0 for col in feature_cols], dtype=float)
    
    best_c = None
    best_logprob = -np.inf
    for c in classes:
        p_j = feature_prob[c]
        # avoid log(0)
        p_j = np.clip(p_j, 1e-12, 1 - 1e-12)
        logprior = np.log(class_prior[c])
        # Bernoulli log-likelihood
        loglik = x * np.log(p_j) + (1 - x) * np.log(1 - p_j)
        total_log = logprior + np.sum(loglik)
        if total_log > best_logprob:
            best_logprob = total_log
            best_c = c
    return best_c

# quick test (optional)
print("Trained NB on", len(classes), "classes. Example prediction:",
      predict_text("predict classification model random forest svm knn"))
```
